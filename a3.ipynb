```python
import jieba
import jieba.analyse
from gensim.models import word2vec
from bs4 import BeautifulSoup
import re
```

```python
#查看文件
with open('three_kingdoms.txt','r',encoding='utf-8') as file:
    for line in file.readlines():
        print(line)
```

```python
#以100字分段
a = 0
b = 0
with open('three_kingdoms.txt','r',encoding='utf-8') as file:
    line = file.read()
    for char in line:
        b+=1
        print(char,end='')
        if b-a == 100:
            a = b
            print('\n')
        if a == 5000:
            break
```

```python
#分词，去标点符号
with open('three_kingdoms.txt',encoding='utf-8') as f:
    document = f.read()
    document = re.sub('[（）：:？“”《》，。！·、\d ]+', ' ', document)  # 去标点
    document_cut = jieba.cut(document) # 结巴分词
    result = ' '.join(document_cut)
    with open('segement.txt', 'w',encoding='utf-8') as f2:
        f2.write(result)
    print('Segement Endding')
```

```python
#查看分词，去标点符号后的文件
a = 0
b = 0
with open('segement.txt','r',encoding='utf-8') as file:
    line = file.read()
    for char in line:
        b+=1
        print(char,end='')
        if b-a == 100:
            a = b
            print('\n')
        if a == 5000:
            break
```

```python
 #训练模型
 sentences = word2vec.LineSentence('segement.txt')
 model = word2vec.Word2Vec(sentences, hs=1,min_count=1,window=3,size=100)
 print('Word2Vec Training Endding!')
 model.save('sanguo.model') #保存模型
```

```python
#查看与曹操相近的词
model =  word2vec.Word2Vec.load('sanguo.model')
print('Nearest 曹操:',model.most_similar('曹操'))
```
